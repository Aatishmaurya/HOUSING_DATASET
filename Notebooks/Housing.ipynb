```python
def cap_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    df[col] = np.where(df[col] > upper_bound, upper_bound,
                       np.where(df[col] < lower_bound, lower_bound, df[col]))
    return df

# Example: Cap price outliers
df = cap_outliers(df, 'price')
```

```python
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns

print("Numeric Columns:", numeric_cols)
print("Categorical Columns:", categorical_cols)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Numeric preprocessing: StandardScaler
numeric_transformer = StandardScaler()

# Categorical preprocessing: One-hot encoding
categorical_transformer = OneHotEncoder(drop='first')

# Combine preprocessors
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

pipe.fit(X_train, y_train)   # ✅ preprocessing is applied automatically
y_pred = pipe.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# --- Models ---
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42, n_estimators=200)
}

# --- Baseline Evaluation ---
results = {}
for name, model in models.items():
    pipe = Pipeline(steps=[('preprocessor', preprocessor),
                           ('model', model)])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    results[name] = {
        "R2": r2_score(y_test, y_pred),
        "RMSE": mean_squared_error(y_test, y_pred),
        "pipeline": pipe
    }

print("\n--- Baseline Results ---")
for name, res in results.items():
    print(f"{name}: R2={res['R2']:.3f}, RMSE={res['RMSE']:.2f}")
```

```python
# --- Feature Importance (from Random Forest) ---
rf_pipe = results["Random Forest"]["pipeline"]
all_features = rf_pipe.named_steps['preprocessor'].get_feature_names_out()
importance = rf_pipe.named_steps['model'].feature_importances_
feat_imp = pd.DataFrame({"feature": all_features, "importance": importance})
feat_imp = feat_imp.sort_values("importance", ascending=False)
print("\n--- Feature Importance (Random Forest) ---")
print(feat_imp.head(15))
```

```python
weak_features = feat_imp[feat_imp["importance"] < 0.01]["feature"].tolist()
print("\nDropping low-importance features:", weak_features)
```

```python
# Filter columns in X
X_new = X.drop(columns=[f.split("__")[1] for f in weak_features if f.startswith("num__")],
               errors="ignore")
```

```python
# --- Retrain best model (Random Forest) ---
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y, test_size=0.2, random_state=42)

rf_pipe_new = Pipeline(steps=[
    ('preprocessor', ColumnTransformer([
        ('num', StandardScaler(), [col for col in numeric_cols if col in X_new.columns]),
        ('cat', OneHotEncoder(drop="first"), categorical_cols)
    ])),
    ('model', RandomForestRegressor(random_state=42, n_estimators=200))
])

rf_pipe_new.fit(X_train_new, y_train_new)
y_pred_new = rf_pipe_new.predict(X_test_new)

print("\n--- Final Model After Feature Selection (Random Forest) ---")
print("R2:", r2_score(y_test_new, y_pred_new))
print("RMSE:", mean_squared_error(y_test_new, y_pred_new))
```

📝 Conclusion

Baseline Models

We tested Linear Regression, Decision Tree, and Random Forest as baseline models.

Linear Regression gave interpretable coefficients but could not capture non-linear relationships well.

Decision Tree performed better by handling non-linearities but tended to overfit.

Random Forest achieved the best balance of accuracy and generalization, making it the preferred model.

Feature Importance & Selection

Using Random Forest’s feature importance, we identified that variables such as area, bathrooms, stories, parking, and furnishing status had strong influence on house prices.

Features like guestroom_num, bath_per_bed, and some categorical indicators contributed very little (<1% importance) and were removed.

After removing weak features, the model’s performance remained stable (R² and RMSE almost unchanged), showing that those features were not necessary.

Final Model

The Random Forest Regressor with selected features was chosen as the final model.

It provided strong predictive performance (high R², low RMSE) and also helped us understand which housing characteristics drive property prices the most.

Key Insights

Bigger area, more bathrooms, and availability of modern amenities (air conditioning, furnishing status) significantly increase house price.

Some features (like redundant ratios or low-variance attributes) do not add value and can be safely dropped.

Feature selection not only simplified the model but also improved interpretability without hurting accuracy.
